```python
import pandas as pd
#train = pd.read_csv("/content/drive/MyDrive/데이콘_전력사용량 예측 대회/train_final.csv")
#test = pd.read_csv("/content/drive/MyDrive/데이콘_전력사용량 예측 대회/test_final.csv")
train = pd.read_csv("태양광처리부탁.csv")
train
```

```python
train_table = train[train['태양광용량(kW)'].notnull()]
train_table= train_table.drop("건물유형",axis=1)
train_table
```

```python
test_table = train[train['태양광용량(kW)'].isnull()]
test_table= test_table.drop(["태양광용량(kW)","건물유형"],axis=1)
test_table
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
import autogluon.core as ag
model = TabularPredictor(label='태양광용량(kW)', eval_metric='mean_absolute_percentage_error')
model.fit(train_table, presets='best_quality', time_limit=1800, num_gpus=1)
preds_solar = model.predict(test_table)
```

```python
test_table['태양광용량(kW)'] = preds_solar
test_table
```
